```python
#1. How many people have diabetes and how many people doesn't have diabetes?",

#getting values
df.Diagnosis.value_counts()
```

```python
#plotting bar chart.

df.Diagnosis.value_counts().plot(kind = 'bar', color=["orchid", 'salmon'])

plt.title("Diabetes count")

plt.xlabel("1 = Diabetes, No Diabetes")

plt.ylabel("Amount");
```

```python
#plottin pie cart
df.Diagnosis.value_counts().plot(kind = 'pie', figsize =(8,6))
plt.legend(['Diabetes','No diabetes'])
```

```python
# '0' represent 'Female'

# '1' represent 'Male'

#GENDER column part

# '0' represent 'No diabetes'

# '1' represent 'Diabetes'

#Target column port

#Now Let's check how many 'Male' and 'Female' ore in the dataset

df.Gender.value_counts()
```

```python
#plottin pie cart
df.Gender.value_counts().plot(kind = 'pie', figsize =(8,6))
plt.title('Male Female Ratio')
plt.legend(['Male','Female'])
```

```python
#2. People of which gender has most heart disease?

pd.crosstab(df.Diagnosis, df.Gender)
```

```python
sns.countplot(x = 'Diagnosis', data= df, hue= 'Gender')
plt.title('Diabetes frequency for Gender')
plt.xlabel('0 = No Diabetes, 1 = Diabetes');
```

```python
#More than 52% male has diabetes and 47% female has diabetes.

#Let's move to question 3

#3. People of which gender has which type of cholestrol?

#counting values for different cholestrol

df.Ethnicity.value_counts()
```

```python
#plotting a bar chart

df.Ethnicity.value_counts().plot(kind = 'bar', color = ['salmon', 'lightskyblue', 'springgreen', 'khaki']) 
plt.title('Cholestrol vs count');
```

```python
pd.crosstab(df.Gender, df.Ethnicity)
```

```python
pd.crosstab(df.Gender, df.Ethnicity).plot(kind= 'bar', color = ['coral', 'lightskyblue', 'plum', 'khaki'] )
plt.title('Type of cholestrol for gender') 
plt.xlabel('0 = Female, 1 = Male');
```
